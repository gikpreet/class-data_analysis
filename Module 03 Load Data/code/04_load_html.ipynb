# HTML 테이블 데이터 로드

```python
import pandas as pd
```

## Load

```python
boxoffice = pd.read_html('https://movie.datamotion.co.kr/Boxoffice')
df_boxoffice = boxoffice[1]
```

```python
df_boxoffice
```

```python
df_boxoffice.columns = df_boxoffice.iloc[0]
df_boxoffice.drop([0], axis=0, inplace=True)
df_boxoffice['개봉연도'] = [x.split('(')[1][0:4] for x in df_boxoffice['영화']]
df_boxoffice.영화 = [x.split('(')[0].strip() for x in df_boxoffice.영화]
df_boxoffice
```
