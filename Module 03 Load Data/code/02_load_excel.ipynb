# Excel 파일 로드

```python
import pandas as pd
```

## Load

```python
df_boxoffice = pd.read_excel('./data/boxoffice.xlsx')
df_boxoffice
```

```python
df_boxoffice.columns = df_boxoffice.iloc[0]
df_boxoffice
```

```python
df_boxoffice.drop(0, axis=0, inplace=True) # axis: 0: row, 1: column
df_boxoffice
```

```python
release_year = [x.split('(')[1][0:4] for x in df_boxoffice['영화']]
release_year
```

```python
df_boxoffice['releaseyear'] = release_year
df_boxoffice
```

```python
df_boxoffice.rename(columns={'releaseyear':'개봉연도'}, inplace=True)
df_boxoffice
```

```python
df_boxoffice.영화 = [x.split('(')[0].strip() for x in df_boxoffice.영화]
df_boxoffice
```

## Save

```python
# openpyxl 라이브러리 필요
# Anaconda 설치: conda install anaconda:openpyxl
# pip 설치: pip install openpyxl
df_boxoffice.to_excel('./data/after_refine.xlsx')
```
