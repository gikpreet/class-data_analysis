# DataFrame 인덱스 사용

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'name':     ['Celine','James','Jason','Hans'],
    'nation':   ['France','UK','USA','Germany'],
    'english':  [90,80,80,60],
    'math':     [100,40,40,50],
    'science':  [100,30,60,90]
})
df
```

```python
df2 = df.set_index('name')
df2
```

```python
df3 = df.set_index(['name', 'nation'])
df3
```

```python
df.index=['a','b','c','d']
```

```python
df4 = df2.reindex(['Celine','Jason', 'James','Robert'])
df4
```

```python
df3.reset_index(inplace=True)
df3
```

```python
df4.sort_index(inplace=True)
df4
```

```python
df5 = df4.copy()
df5
```
