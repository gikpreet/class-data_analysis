```python
import pandas as pd
```

## pivot_table() 메소드

```python
titanic = pd.read_csv('titanic.csv')
```

```python
df_객실등급별_남녀의나이_중간값 = pd.pivot_table(titanic,
                    index = 'Pclass',    # 행 위치에 들어갈 열
                    columns = 'Sex',    # 열 위치에 들어갈 열
                    values = 'Age',     # 데이터로 사용할 열
                    aggfunc = 'mean')   # 데이터 집계함수
```

```python
df_객실등급별_남녀의나이_중간값
```

```python
df_객실등급별_남녀의숫자 = pd.pivot_table(titanic,
                    index = 'Pclass',    # 행 위치에 들어갈 열
                    columns = 'Sex',    # 열 위치에 들어갈 열
                    values = 'Age',     # 데이터로 사용할 열
                    aggfunc = 'count')   # 데이터 집계함수
```

```python
df_객실등급별_남녀의숫자
```

```python
df_객실등급별_남녀생존자 = pd.pivot_table(titanic,
                     index = ['Pclass','Sex'],
                     columns = 'Survived',
                     values = ['Age','Fare'],
                     aggfunc = ['mean','max'])
```

```python
df_객실등급별_남녀생존자
```
