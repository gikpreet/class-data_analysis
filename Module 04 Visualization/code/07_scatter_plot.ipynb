# Scatter plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 한글 처리
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
df_titanic = sns.load_dataset('titanic')
```

```python
df_titanic.plot(kind='scatter',x='age',y='fare',s=10)
```

```python
sns.scatterplot(data=df_titanic, x='age', y='fare', hue='survived', palette='coolwarm')
```

## 자동차 데이터

```python
df_car = pd.read_csv('./data/auto-mpg.csv')
df_car.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
df_car.head()
```

```python
df_car.plot(kind='scatter',x='weight',y='mpg',c='blue',s=10,figsize=(10,5))
plt.title('Scatter plnt = 연비 vs. 무게')
plt.show()
```
